# Conservation's Laws

```python
# Angular Momentum
def angular_momentum(x, y, z, vx, vy, vz):
    lx = y*vz - z*vy
    ly = z*vx - x*vz
    lz = x*vy - y*vx
    return lx, ly, lz

# Energy
def energy(vx, vy, vz, u):
    return 0.5*(vx*vx + vy*vy + vz*vz) + u
```

```python
T = 3
time = range(T+1)
ang_mom = np.zeros(T+1)
lin_mom = np.zeros(T+1)
energ = np.zeros(T+1)

for ii in range(T+1):
    df = src.read_csv('../Data/results/sedov_blast_wave/' + str(ii) + ".csv")
    L = np.sum(angular_momentum(df['x'], df['y'], df['z'], df['vx'], df['vy'], df['vz']), axis=1)
    ang_mom[ii] = np.sqrt(L[0]*L[0] + L[1]*L[1] + L[2]*L[2])
    vx = np.sum(df['vx'])
    vy = np.sum(df['vy'])
    vz = np.sum(df['vz'])
    lin_mom[ii] = np.sqrt(vx*vx + vy*vy + vz*vz)
    energ[ii] = np.sum(energy(df['vx'], df['vy'], df['vz'], df['u']))
```

```python
plt.plot(time[:], np.abs(lin_mom[:]-lin_mom[0]), '-', color='blue')
plt.title("Conservation of Linear Momentum")
plt.xlabel("it [x50]")
plt.ylabel(r"$\frac{|P-P_0|}{m}$")
plt.xlim(0, T)
plt.show()
```

```python
plt.plot(time[:], np.abs(ang_mom[:]-ang_mom[0]), '-', color='crimson')
plt.title("Conservation of Angular Momentum")
plt.xlabel("it [x50]")
plt.ylabel(r"$\frac{|L-L_0|}{m}$")
plt.xlim(0, T)
plt.show()
```

```python
plt.plot(time[:], np.abs((energ[:]-energ[0])/energ[0]), '-', color='forestgreen')
plt.title("Conservation of Energy")
plt.xlabel("it [x50]")
plt.ylabel(r"$\left|\frac{E-E_0}{E_0}\right|$")
plt.xlim(0, T)
plt.show()
```

# Density distribution

```python
t = 0.08082978661933742
wf = 1.1527*(t)**(0.4)
dm32 = 1/32**3
dm64 = 1/64**3
```

```python
sdf32 = src.read_csv('../Data/results/sedov_blast_wave/final32K.csv')
rad32 = np.sqrt((sdf32["x"]-0.5)**2 + (sdf32["y"]-0.5)**2+(sdf32["z"]-0.5)**2)
sdf32.params={'mass':dm32, 'hfact':eta}
sdf32.calc_density()
sdf64 = src.read_csv('../Data/results/sedov_blast_wave/final64K.csv')
rad64 = np.sqrt((sdf64["x"]-0.5)**2 + (sdf64["y"]-0.5)**2+(sdf64["z"]-0.5)**2)
sdf64.params={'mass':dm64, 'hfact':eta}
sdf64.calc_density()
```

```python
fig = plt.figure(figsize=(8,8))
plt.plot(rad32, sdf32['rho'], ".", color="crimson", markersize=1, label="32K particles")
plt.plot(rad64, sdf64['rho'], ".", color="blue", markersize=1, label = "216K particles")
plt.axvline(x=wf, ls='--', c='k')
plt.xlabel("r", fontsize=12)
plt.ylabel(r"$\rho$", fontsize=12)
plt.legend(markerscale=8, fontsize=12)
plt.savefig("SedovSPH.png")
plt.show()
```
